```python
listquotes = []

# Break down the BS4 object manually and put all the information of interest in a standard object
for q in quotesoup:
    text_quote = q.find("span",{"class":"text"}).text 
    author_name = q.find("small" , {"class":"author"}).text
    
    #tag_links returns <a ... /a>, so need addtional procedure to extract juss the tag
    tag_links = q.find_all("a", {"class":"tag"})
    tags = []
    
    for t in tag_links:
        tags.append(t.text)
    
    
    # Pack everything into an object
    quote_object = {
        "quote":text_quote,
        "author":author_name,
        "tags":tags
    }
    
    # Put object into list, so Dataframes can consume it and not shout back at me
    listquotes.append(quote_object)

# Et Voil'a
pd.DataFrame(listquotes)


    
```
